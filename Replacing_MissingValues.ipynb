### Les techniques de remplacement de base

```python
# Par une constante : 
df['ST_NUM'].fillna(125, inplace = True)
```

```python
df
```

```python
# Par une valeur médiane (fréquemment utilisée):
median = df['NUM_BEDROOM'].median()
df['NUM_BEDROOM'].fillna(median, inplace=True)
#inplace--> affichage complet du tableau
```

```python
df
```

```python
# Remplir par la valeur précédente (forward-fill)
df.NUM_BEDROOM.fillna(method='ffill',inplace=True)
df
```

```python
# Remplir par la valeur suivante (back-fill)
df.NUM_BEDROOM.fillna(method='bfill')
```
